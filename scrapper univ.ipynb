```python
# Inisialisasi list untuk menampung data
data = []


# Temukan semua baris <tr> dalam <tbody>
rows = reviews.find_all('tr')

# Looping untuk mengambil data dari setiap baris
for row in rows:
    # Ambil semua <td> dalam setiap baris
    cols = row.find_all('td')
    
    # Pastikan ada cukup kolom sebelum mengekstrak data
    if len(cols) >= 4:
        # Ambil data dari setiap kolom
        program_studi = cols[0].get_text(strip=False)
        jenjang = cols[1].get_text(strip=False)
        lembaga = cols[2].get_text(strip=False)
        link_aksi = cols[3].find('p').get_text(strip=False)
        
        # Masukkan data ke dalam list
        data.append([program_studi, jenjang, lembaga, link_aksi])

# Buat DataFrame
df = pd.DataFrame(data, columns=["Program Studi", "Jenjang", "Lembaga", "Aksi"])

# Tampilkan DataFrame
df
```

```python
# Parse HTML
soup = reviews.find_all('div', class_='space-y-1')[1]
soup

# Ekstraksi hanya nomor telepon pertama

# Ekstraksi data kontak
first_phone = soup.find('p', class_='font-semibold text-l flex').get_text(strip=True)
email = [p.get_text(strip=True) for p in soup.find_all('p', class_='font-semibold text-l flex') if '@' in p.get_text()]
website = [p.get_text(strip=True) for p in soup.find_all('p', class_='font-semibold text-l flex') if 'http' in p.get_text()]

# Menyusun data ke dalam dictionary
data = {
    'Phone': first_phone,
    'Email': email,
    'Website': website
}

# Membuat DataFrame dari dictionary
df2 = pd.DataFrame(data)

# Tampilkan DataFrame
df2
```

```python
# Parse HTML
soup = reviews

# Ekstraksi data berdasarkan kelas yang sesuai
kode = soup.find('p', string='Kode').find_next('p').get_text(strip=True)
status = soup.find('p', string='Status').find_next('p').get_text(strip=True)
akreditasi = soup.find('p', string='Akreditasi').find_next('p').get_text(strip=True)
biaya = soup.find('p', string='Biaya Kuliah').find_next('p').get_text(strip=True)
indikator_kelengkapan = soup.find('p', string='Indikator Kelengkapan Data').find_next('p').get_text(strip=True)
tanggal_berdiri = soup.find('p', string='Tanggal Berdiri').find_next('p').get_text(strip=True)
no_sk = soup.find('p', string='No SK Pendirian').find_next('p').get_text(strip=True)
tanggal_sk = soup.find('p', string='Tanggal SK Pendirian').find_next('p').get_text(strip=True)


# Menyusun data ke dalam bentuk dictionary
data = {
    'Kode': [kode],
    'Status': [status],
    'Akreditasi': [akreditasi],
    'Biaya Kuliah': [biaya],
    'Indikator Kelengkapan Data': [indikator_kelengkapan],
    'Tanggal Berdiri': [tanggal_berdiri],
    'NO SK': [no_sk],
    'Tanggal SK': [tanggal_sk]
}

# Membuat DataFrame dari dictionary
df = pd.DataFrame(data)

# Tampilkan DataFrame
df
```
